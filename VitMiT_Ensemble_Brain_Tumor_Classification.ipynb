```python
import pandas as pd
import numpy as np

data = {
    "Employee_ID": [101,102,103,104,105],
    "Salary": [50000, 70000, 69000, 20000, np.nan],
    "Join_Date": ["2022/05/01", "2021/12/19", "2021/12/21", "2021/12/28", "2021/12/15"],
    "Department": ["HR", "Human Resources", "hr", "HR ", "human resources"]
}

df = pd.DataFrame(data)

df
```

```python
#findi=ing anomalies
df["Salary"].describe()
```

```python
#sus values frequency

df["Salary"].value_counts()
```

```python


df["Salary"] = df["Salary"].replace([999999, -1], np.nan)


df["Salary"].fillna(df["Salary"].median(), inplace=True)

df
```

```python
#cleaning and standardization
df["Department"] = df["Department"].str.lower().str.strip()

df["Department"] = df["Department"].replace({
    "hr": "Human Resources",
    "human resources": "Human Resources"
})#standardize the string hr and human resources, to one single Human Resources

df
```

```python


df["Join_Date"] = pd.to_datetime(df["Join_Date"], errors="coerce")

df
```

```python
#orders dataset

orders_data = {
    "Order_ID": [1,2,3,4],
    "Product_ID": [101,102,101,103],
    "Quantity": [2, np.nan, 5, 3]
}

DF_Orders = pd.DataFrame(orders_data)

DF_Orders
```

```python
#product info dataset

product_data = {
    "Product_ID": [101,102,103],
    "Product_Name": ["Laptop", "Mouse", "Keyboard"],
    "Price": [80000, 5000, 2500]
}

DF_Product_Info = pd.DataFrame(product_data)

DF_Product_Info
```

```python
#Backup dataset, to fill in the missing vals
DF_Orders_Backup=pd.DataFrame({
    "Order_ID": [1, 2, 3],
    "Product_ID": [101, 102, 103],
    "Quantity": [2, 1, 5]})
```

```python
#merge orders and product info dataframes

merged_df = pd.merge(DF_Orders, DF_Product_Info, on="Product_ID")

merged_df
```

```python
#Sales = Quantity × Price (given)

merged_df["Sales"] = merged_df["Quantity"] * merged_df["Price"]

merged_df
```

```python
pivot_table = merged_df.pivot_table(
    index="Product_Name",
    values="Sales",
    aggfunc="sum"
)
print(pivot_table)
```

```python
#Using combine_first to update missing vals w/o duplicating rows

DF_Orders_Backup = DF_Orders_Backup.set_index("Order_ID")

DF_Orders["Quantity"] = DF_Orders["Quantity"].combine_first(
    DF_Orders_Backup["Quantity"]
)

DF_Orders = DF_Orders.reset_index()

print(DF_Orders)
```
